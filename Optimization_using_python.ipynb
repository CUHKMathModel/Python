# Solving Unconstrained and Constrained Optimization Problems in Python


```python
import numpy as np
from scipy.optimize import minimize, Bounds
```

## 1. Unconstrained Optimization

An **unconstrained optimization** problem seeks to find the minimum (or maximum) of an objective function $f(\mathbf{x})$ without restrictions on $\mathbf{x}$:

$$\min_{\mathbf{x}} f(\mathbf{x})$$

Main function: **scipy.optimize.minimize**

Different numerical methods for solving the optimization problem: can choose among  'BFGS', 'L-BFGS-B', 'Nelder-Mead', 'SLSQP'

### Example 1: Minimize $f(x) = (x-3)^2 + (y+1)^2$

```python
# Define target functions 定義目標函數（input is a vector 輸入為向量 x = [x, y]）
def f_quad(x):
    return (x[0] - 3)**2 + (x[1] + 1)**2

# Analytical minimum
print("Analytical minimum at x=[3, -1], f=0")

# Using scipy.optimize.minimize (default: BFGS for unconstrained)
res = minimize(f_quad, [0, 0], method='BFGS')
print("Numerical solution:")
print("x* =", res.x)
print("f(x*) =", res.fun)
print("Success:", res.success)
```

### Example 2: Minimize $f(x,y,z) = x^2 + 2y^2 + 3z^2 + 2xy + 2xz + 4yz$

```python
# define target functions 定義目標函數（input is a vector 輸入為向量 x = [x, y, z]）
def f(vars):
    x, y, z = vars
    return x**2 + 2*y**2 + 3*z**2 + 2*x*y + 2*x*z + 4*y*z

# initial guess (important!) 初始猜測值（很重要！）
initial_guess = [1.0, 1.0, 1.0]

# Perform optimization 執行優化
result = minimize(f, initial_guess, method='BFGS')

print("Minimum point location 最小值位置:", result.x)
print("Minimum value 最小值 f(x):", result.fun)
```

## 2. Constrained Optimization

A **constrained optimization** problem seeks to find the minimum (or maximum) of an objective function $f(\mathbf{x})$ with certain constraints:

$$\min_{\mathbf{x}} f(\mathbf{x})$$

subject to

**Equality constraints**: $h_i(\mathbf{x}) = 0$

**Inequality constraints**: $g_j(\mathbf{x}) \leq 0$


### Example 3: Minimize $f(x,y) = x^2 + y^2$ subject to $x+y=1$

```python
def objective(x):
    return x[0]**2 + x[1]**2  

def eq_constraint(x):
    return x[0] + x[1] - 1  # x + y = 1  => x+y-1 = 0 (we always need RHS to be 0)

# set up the equality constraint
cons = {'type': 'eq', 'fun': eq_constraint}

res = minimize(objective, [0.5, 0.5], constraints=cons, method='SLSQP')
print("Solution:", res.x)
print("f(x*):", res.fun)
```

### Example 4: Maximize $f(x,y) = x+y$ subject to $x^2 + y^2 \leq 1$

```python
def obj_ineq(x):
    return -(x[0] + x[1])  # maximize x+y => minimize -(x+y)

def ineq1(x):
    return x[0]**2 + x[1]**2 - 1  # inside unit circle x^2 + y^2 <=1

# set up the inequality constraint
cons_ineq = {'type': 'ineq', 'fun': ineq1}

bounds = Bounds([-1,-1], [1,1])

res = minimize(obj_ineq, [0, 0], constraints=cons_ineq, bounds=bounds, method='SLSQP')
print("Solution:", res.x)
print("Max x+y:", -res.fun)
```

### Example 5: An integrated example

A factory produces three types of products with quantities (say, in tons) $  x, y, z  $ respectively, and the goal is to minimize the cost:
某工廠生產三種產品 $  x, y, z  $，目標是最小化成本：

$$   f(x, y, z) = 4x^2 + 3y^2 + 2z^2 + 5xy + 3xz   $$

Constraints 限制條件：

* Equality constraint (total production = 10) 等式約束（總產量必須為 10）：$  x + y + z = 10  $
* Inequality constraints 不等式約束：
  * $  x + y \geq 6  $ 
  * $  yz \leq 20  $ 
  * $  x \leq 8  $ 

```python
def objective(x):
    x1, x2, x3 = x
    return 4*x1**2 + 3*x2**2 + 2*x3**2 + 5*x1*x2 + 3*x1*x3

# Equality: x + y + z = 10
def eq_cons(x):
    return x[0] + x[1] + x[2] - 10

# Inequality:
cons = [
    {'type': 'eq',   'fun': eq_cons},                    # Equality: x + y + z = 10
    {'type': 'ineq', 'fun': lambda x: x[0] + x[1] - 6},  # x+y >= 6 (note: lambda x: ... is another way to define the function inline)
    {'type': 'ineq', 'fun': lambda x: 20 - x[1]*x[2]}    # yz <= 20
]

# bounds
bounds = [(0, 8), (0, None), (0, None)]

x0 = [3, 3, 4]

result = minimize(objective, x0, method='SLSQP', bounds=bounds, constraints=cons)

print("Success: ", result.success)
print("x*:", result.x)
print("Minimum cost", result.fun)
```
