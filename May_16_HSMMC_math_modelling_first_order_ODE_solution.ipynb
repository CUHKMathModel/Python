# Example 1: Forward Euler Formula

**Task 1.1: Implement the forward Euler scheme to solve the ODE**

To do: replace `...` below to complete the task

Tasks:
+  Implements the forward Euler scheme (which works for growth ($k>0$) and decay ($k<0$))
+  Compares numerical results with the exact analytical solution ($y(t) = y_0 e^{kt}$)
+  Plots both solutions for visualization

ODE:
$$
\frac{dy}{dt} = ky,\quad y(0) = y_0
$$

**Forward Euler Formula**:
$$
y_{n+1} = y_n + dt \cdot k \cdot y_n
$$

where

+ $y_n$ = current value
+ $dt$ = time step
+ $y_{n+1}$ = next value

Set
* initial value $y_0 = 100$
* rate constant $k=0.1$
* total simulation time $t_{\text end} = 20$
* time step $dt = 0.5$

Hint:
* `np.arrange(start, stop, step)` returns an array with values generated within the half-open interval $[$ `start` , `stop` $)$ , with spacing between values given by `step`.
* `len(x)` returns the length of the array `x`
* `np.zeros(n)` returns an aray with values $0$ of size `n`

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Step 1.0: User-defined parameters
# --------------------------
y0 = 100        # Initial value (e.g., population, mass)
k = 0.1         # Rate constant: k>0 = growth, k<0 = decay
t_end = 20      # Total simulation time
dt = 0.5        # Time step (smaller = more accurate)

# --------------------------
# Step 1.1: Create time array
# --------------------------
t = np.arange(0, t_end + dt, dt)
n_steps = len(t)

# --------------------------
# Step 1.2: Forward Euler (Forward Scheme)
# --------------------------
y_euler = np.zeros(n_steps)
y_euler[0] = y0  # Set initial condition

# Iterate using forward scheme
for i in range(n_steps - 1):
    y_euler[i+1] = y_euler[i] + dt * k * y_euler[i]

# --------------------------
# Step 1.3: Exact analytical solution
# --------------------------
y_exact = y0 * np.exp(k * t)

# --------------------------
# Step 1.4: Calculate half-life (for decay only)
# --------------------------
if k < 0:
    half_life = np.log(2) / abs(k)
    print(f"=== Radioactive Decay ===")
    print(f"Half-life T = {half_life:.2f} time units")
else:
    print(f"=== Exponential Growth ===")
```

**Task 1.2: Plot the results to compare numerical results with the exact analytical solution**

```python
# --------------------------
# Step 1.5: Plot results
# --------------------------
plt.figure(figsize=(10, 5))
plt.plot(t, y_euler, 'o-', label='Forward Euler (Numerical)', linewidth=2, markersize=4)
plt.plot(t, y_exact, '-', label='Exact Solution (Analytical)', linewidth=2)
plt.xlabel('Time t')
plt.ylabel('Quantity y(t)')
plt.title(f'Exponential Change (k={k}, y0={y0})')
plt.legend()
plt.grid(True)
plt.show()

# --------------------------
# Step 1.6: Print first 5 values for verification
# --------------------------
print("\nFirst 5 time points:")
print("t\tEuler\t\tExact")
for i in range(5):
    print(f"{t[i]:.1f}\t{y_euler[i]:.2f}\t\t{y_exact[i]:.2f}")
```

# Example 2: Backward Euler Formula

**Task 2.1: Implement the backward Euler scheme to solve the ODE**

To do: replace `...` below to complete the tasks

Tasks:

+ Implement the backward Euler scheme (which works for exponential growth ($k>0$) and decay ($k<0$))
+ Compare with the exact analytical solution
+ Plot the results

ODE:
$$
\frac{dy}{dt} = ky,\quad y(0) = y_0
$$

**Backward Euler Formula**
\begin{align}
y_{n+1} &= y_n + dt\cdot k\cdot y_{n+1} \\
\Longrightarrow \quad y_{n+1} &= \frac{y_n}{1 - k\cdot dt}
\end{align}

Set
* initial value $y_0 = 100$
* rate constant $k=0.1$
* total simulation time $t_{\text end} = 20$
* time step $dt = 0.5$

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Step 2.0 Simulation Parameters
# --------------------------
y0 = 100        # Initial value
k = 0.1         # Rate constant (k>0=growth, k<0=decay)
t_end = 20      # Total simulation time
dt = 0.5        # Time step

# --------------------------
# Step 2.1: Time array
# --------------------------
t = np.arange(0, t_end + dt, dt)
n_steps = len(t)

# --------------------------
# Step 2.2: Backward Euler (Backward Scheme)
# --------------------------
y_backward = np.zeros(n_steps)
y_backward[0] = y0

# Iterate with backward Euler formula
for i in range(n_steps - 1):
    # Rearranged explicit formula for linear ODE dy/dt = ky
    y_backward[i+1] = y_backward[i] / (1 - k * dt)

# --------------------------
# Step 2.3: Exact Analytical Solution
# --------------------------
y_exact = y0 * np.exp(k * t)

# --------------------------
# Step 2.4: Half-life calculation (for decay)
# --------------------------
if k < 0:
    half_life = np.log(2) / abs(k)
    print("=== Exponential Decay ===")
    print(f"Half-life T = {half_life:.2f} time units")
else:
    print("=== Exponential Growth ===")
```

**Task 2.2: Plot the results to compare numerical results with the exact analytical solution**

```python
# --------------------------
# Step 2.5: Plot Results
# --------------------------
plt.figure(figsize=(10, 5))
plt.plot(t, y_backward, 's-', label='Backward Euler (Numerical)', linewidth=2, markersize=5)
plt.plot(t, y_exact, '-', label='Exact Solution', linewidth=2)
plt.xlabel('Time t')
plt.ylabel('Quantity y(t)')
plt.title(f'Backward Euler - Exponential Change (k={k}, y0={y0})')
plt.legend()
plt.grid(True)
plt.show()

# --------------------------
# Step 2.6: Print sample values
# --------------------------
print("\nFirst 5 time points comparison:")
print(f"t\tBackward Euler\tExact")
for i in range(5):
    print(f"{t[i]:.1f}\t{y_backward[i]:.2f}\t\t{y_exact[i]:.2f}")
```

# Example 3: Central Difference Scheme

**Task 3.1: Implement the Central Difference Scheme to solve the ODE**

To do: replace `...` below to complete the tasks

Tasks:

+ Implement the backward Euler scheme (which works for exponential growth ($k>0$) and decay ($k<0$))
+ Compare with the exact analytical solution
+ Plot the results

ODE:
$$
\frac{dy}{dt} = ky,\quad y(0) = y_0
$$


**Central difference scheme (2nd-order accurate)**:

\begin{align}
 \frac{y_{n+1} - y_{n-1}}{2dt} &= k y_n. \\
\Longrightarrow  \quad \quad \quad \quad y_{n+1} &= y_{n-1} + 2\cdot dt\cdot k\cdot y_n.
\end{align}

* Needs two initial values: $y_0$ and $y_1$.
* We use Forward Euler to compute $y_1$ for initialization.

Set
* initial value $y_0 = 100$
* rate constant $k=0.1$
* total simulation time $t_{\text end} = 20$
* time step $dt = 0.5$

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Step 3.0: Simulation Parameters
# --------------------------
y0 = 100        # Initial value y(0)
k = 0.1        # Rate constant (k>0=growth, k<0=decay)
t_end = 20      # Total simulation time
dt = 0.5        # Time step

# --------------------------
# Step 3.1: Create time array
# --------------------------
t = np.arange(0, t_end + dt, dt)
n_steps = len(t)

# --------------------------
# Step 3.2: Central Difference Scheme (Leapfrog)
# --------------------------
y_central = np.zeros(n_steps)
y_central[0] = y0  # Initial condition

# Step 3.2.1: Compute y1 using Forward Euler (initialization for multi-step method)
y_central[1] = y_central[0] + dt * k * y_central[0]

# Step 3.2.2: Central difference update for all remaining steps
for n in range(1, n_steps - 1):
    y_central[n+1] = y_central[n-1] + 2 * dt * k * y_central[n]

# --------------------------
# Step 3.3: Exact Analytical Solution
# --------------------------
y_exact = y0 * np.exp(k * t)

# --------------------------
# Step 3.4: Half-life (for decay only)
# --------------------------
if k < 0:
    half_life = np.log(2) / abs(k)
    print("=== Exponential Decay ===")
    print(f"Half-life T = {half_life:.2f} time units")
else:
    print("=== Exponential Growth ===")
```

**Task 3.2: Plot the results to compare numerical results with the exact analytical solution**

```python
# --------------------------
# Step 3.5: Plot Results
# --------------------------
plt.figure(figsize=(10, 5))
plt.plot(t, y_central, 'D-', label='Central Difference Scheme', linewidth=2, markersize=5)
plt.plot(t, y_exact, '-', label='Exact Solution', linewidth=2)
plt.xlabel('Time t')
plt.ylabel('Quantity y(t)')
plt.title(f'Central Difference - Exponential Change (k={k}, y0={y0})')
plt.legend()
plt.grid(True)
plt.show()

# --------------------------
# Step 3.6: Print comparison table
# --------------------------
print("\nFirst 5 time points:")
print(f"t\tCentral\t\tExact")
for i in range(5):
    print(f"{t[i]:.1f}\t{y_central[i]:.2f}\t\t{y_exact[i]:.2f}")
```

# Example 4: A Salt Water Mixing Problem

$$
\frac{dA}{dt} + \frac{r_2}{(r_1 - r_2)t + V_0} A(t) = c_1 r_1
$$

**Task 4.1: Implement the Central Difference Scheme to solve the ODE**

To do: replace `...` below to complete the tasks

Tasks:

+ Implement the Central Difference scheme (which works for exponential growth ($k>0$) and decay ($k<0$))
+ Compare with the exact analytical solution
+ Plot the results

ODE:
$$
\frac{dA}{dt} = 10 - \frac{4A}{100+t}, \quad A(0)=50
$$

Analytic solution:
$$ A(t) = 2(100 + t) - \frac{150 \cdot 100^4}{(100 + t)^4} $$

**Central Difference Formula**:

$$
A_{i+1} = A_{i-1} + 2\Delta t\cdot \frac{dA}{dt}\left(t_i,A_i\right)
$$

+ Multi-step method: needs $A_0$ and $A_1$
+ $A_1$ computed with forward Euler
+ 2nd-order accuracy $\to$ very close to exact solution

**Example Parameters**:

+ $A_0 = 50$ lb
+ $V_0 = 100$ gal
+ $r_1 = 5,\ r_2 = 4$ gal/min
+  $c_1 = 2$ lb/gal
+  Solve up to $t=25$ min
+ time step: $dt = 0.1$ min

**Volume function**:
$$V(t) = V_0 +(r_1 - r_2)t$$

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Step 4.0: PARAMETERS (MATCH YOUR PROBLEM)
# --------------------------
A0 = 50          # Initial salt amount (lb)
V0 = 100         # Initial volume (gal)
r1 = 5           # Inflow rate (gal/min)
r2 = 4           # Outflow rate (gal/min)
c1 = 2           # Inlet concentration (lb/gal)
t_end = 25       # Simulate up to 25 minutes
dt = 0.1         # Time step (small = accurate)

# --------------------------
# Step 4.1: Time array
# --------------------------
t = np.arange(0, t_end + dt, dt)
n = len(t)

# --------------------------
# Step 4.2: Volume function V(t) (analytical)
# --------------------------
def volume(t):
    return V0 + (r1 - r2) * t

# --------------------------
# Step 4.3: ODE Function: dA/dt = 10 - (4*A)/(100 + t)
# --------------------------
def dA_dt(t_val, A_val):
    V = volume(t_val)
    return c1*r1 - (r2 * A_val) / V

# --------------------------
# Step 4.4: CENTRAL DIFFERENCE (LEAPFROG) SCHEME
# --------------------------
A_central = np.zeros(n)
A_central[0] = A0  # Initial condition

# Step 4.4.1: Compute A1 with FORWARD EULER (initialize multi-step method)
A_central[1] = A_central[0] + dt * dA_dt(t[0], A_central[0])

# Step 4.4.2: Central difference for all remaining steps
for i in range(1, n-1):
    # Leapfrog formula: A_{i+1} = A_{i-1} + 2*dt*dA/dt(t_i, A_i)
    A_central[i+1] = A_central[i-1] + 2 * dt * dA_dt(t[i], A_central[i])

# --------------------------
# Step 4.5: EXACT ANALYTICAL SOLUTION (for comparison)
# --------------------------
def exact_solution(t_val):
    V = volume(t_val)
    C = -150 * (100**4)
    return 2 * V + C / (V**4)

A_exact = exact_solution(t)

# --------------------------
# Step 4.6: RESULTS AT t=25 min
# --------------------------
t_target = 25
idx = np.where(np.isclose(t, t_target))[0][0]

V25 = volume(t_target)
A25_central = A_central[idx]
A25_exact = exact_solution(t_target)
conc_central = A25_central / V25
conc_exact = A25_exact / V25

# --------------------------
# Step 4.7: PRINT OUTPUT
# --------------------------
print("="*60)
print("MIXING PROBLEM - CENTRAL DIFFERENCE vs EXACT SOLUTION")
print("="*60)
print(f"Time = {t_target} min")
print(f"Volume V(t) = {V25:.0f} gal")
print(f"\nSalt Amount A(t):")
print(f"Central Difference: {A25_central:.2f} lb")
print(f"Exact Solution:     {A25_exact:.2f} lb")
print(f"\nConcentration:")
print(f"Central Difference: {conc_central:.3f} lb/gal")
print(f"Exact Solution:     {conc_exact:.3f} lb/gal")
print("="*60)
```

**Task 4.2: Plot the results to compare numerical results with the exact analytical solution**

```python
# --------------------------
# Step 4.8: PLOT
# --------------------------
plt.figure(figsize=(10,5))
plt.plot(t, A_central, 'o-', label='Central Difference', markersize=2, linewidth=1.5)
plt.plot(t, A_exact, '-', label='Exact Solution', linewidth=2)
plt.xlabel('Time t (minutes)')
plt.ylabel('Amount of Salt A(t) (lb)')
plt.title('Mixing Tank Problem: Central Difference Scheme')
plt.legend()
plt.grid(True)
plt.show()
```

# Example 5: Fertilizer Mixing Problem with a Maximum

**Task 5.1: Implement the Central Difference Scheme to solve the ODE**

To do: replace `...` below to complete the tasks

Tasks:

+ Implement the Central Difference scheme (which works for exponential growth ($k>0$) and decay ($k<0$))
+ Compare with the exact analytical solution
+ Plot the results

ODE:

$$\frac{dA}{dt} = 1 - \frac{3A}{100-2t}, \quad A(0)=0$$

Analytic solution:

$$A(t) = (100 - 2t) - \frac{1}{10}(100 - 2t)^{3/2}$$

**Central Difference Update Rule:**

$$
A_{i+1} = A_{i-1} + 2\Delta t \cdot \frac{dA}{dt}\left(t_i, A_i\right)
$$

**Example Parameters**:

+ $A_0 = 0$ lb
+ $V_0 = 100$ gal
+ $r_1 = 1,\ r_2 = 3$ gal/min
+  $c_1 = 1$ lb/gal
+  Solve up to $t=49.9$ min
+ time step: $dt = 0.01$ min

**Volume function**:
$$V(t) = V_0 +(r_1 - r_2)t$$


Hint:
* `np.max(x)` returns the maximum element in the array `x`
* `np.argmax(x)` returns the index where the maximum element is in the array `x`

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Step 5.0: Problem Parameters
# --------------------------
A0 = 0            # Initial fertilizer (lb)
V0 = 100          # Initial volume (gal)
r1 = 1            # Inflow rate (gal/min)
r2 = 3            # Outflow rate (gal/min)
c1 = 1            # Inlet concentration (lb/gal)
t_end = 49.9      # Avoid t=50 (empty tank)
dt = 0.01         # Small time step for accuracy

# --------------------------
# Step 5.1: Time array
# --------------------------
t = np.arange(0, t_end + dt, dt)
n_steps = len(t)

# --------------------------
# Step 5.2: Volume function V(t)
# --------------------------
def volume(t_val):
    return V0 + (r1 - r2) * t_val

# --------------------------
# Step 5.3: ODE: dA/dt = 1 - (3A)/(100 - 2t)
# --------------------------
def dA_dt(t_val, A_val):
    V = volume(t_val)
    return 1 - (3 * A_val) / V

# --------------------------
# Step 5.4: CENTRAL DIFFERENCE (LEAPFROG) SCHEME
# --------------------------
A_central = np.zeros(n_steps)
A_central[0] = A0

# Initialize first step with Forward Euler (required for multi-step)
A_central[1] = A_central[0] + dt * dA_dt(t[0], A_central[0])

# Central difference loop
for i in range(1, n_steps - 1):
    A_central[i+1] = A_central[i-1] + 2 * dt * dA_dt(t[i], A_central[i])

# --------------------------
# Step 5.5: EXACT ANALYTICAL SOLUTION
# --------------------------
def exact_A(t_val):
    V = volume(t_val)
    return V - (1/10) * (V ** 1.5)

A_exact = exact_A(t)

# --------------------------
# Step 5.6: Find Maximum (from numerical solution)
# --------------------------
max_A = np.max(A_central)
t_max = t[np.argmax(A_central)]

# --------------------------
# Step 5.7: Print Results
# --------------------------
print("=" * 70)
print("FERTILIZER MIXING PROBLEM - CENTRAL DIFFERENCE")
print("=" * 70)
print(f"Maximum fertilizer amount: {max_A:.2f} lb")
print(f"Time at maximum:          {t_max:.2f} min")
print("\nCompare to analytical:")
print(f"Analytical max:    14.81 lb")
print(f"Analytical time:   27.78 min")
print("=" * 70)
```

**Task 5.2: Plot the results to compare numerical results with the exact analytical solution**

```python
# --------------------------
# Step 5.8: Plot
# --------------------------
plt.figure(figsize=(10, 5))
plt.plot(t, A_central, '.-', label='Central Difference', markersize=1)
plt.plot(t, A_exact, label='Exact Solution', linewidth=2)
plt.axvline(t_max, color='red', linestyle='--', label=f'Max at t={t_max:.2f} min')
plt.xlabel('Time t (min)')
plt.ylabel('Amount of Fertilizer A(t) (lb)')
plt.title('Central Difference Scheme - Fertilizer Mixing Problem')
plt.legend()
plt.grid(True)
plt.show()
```
