# Orthogonal Projection and Least Squares Problem


* The least-squares method is the most intuitive way to obtain the best fit curve representing the data patterns.


# Question 1: Projection and least-squares solution

* Let's consider the problem of finding $t$ satisfying the following:
$$
\min \| t\mathbf{a} - \mathbf{x} \|
$$
where $t$ is a real number.

* In other words, this is the problem of finding $t$ that minimizes the distance between the vector $\mathbf{x}$ and the straight line containing $\mathbf{a}$, where $t$ is the scalar.

* We can see that $\| t\mathbf{a} - \mathbf{x} \|$ represents the distance between $t\mathbf{a}$ and $\mathbf{x}$ as shown in Figure 2.
* Intuitively, the shortest distance can be obtained when $\mathbf{p} = t\mathbf{a}$ and $(\mathbf{x} - t\mathbf{a}) \perp \mathbf{a}$.
* Such $t$ is a solution for $\min \| t\mathbf{a} - \mathbf{x} \|$, and the vector $\mathbf{p} = t\mathbf{a}$ is called the projection of $\mathbf{x}$ onto $\mathbf{a}$.

Since $(\mathbf{x} - t\mathbf{a}) \perp \mathbf{a}$, this $t$ can be obtained as follows:

$$
\mathbf{a} \cdot (\mathbf{x} - t\mathbf{a}) = 0 \implies t(\mathbf{a} \cdot \mathbf{a}) = \mathbf{a} \cdot \mathbf{x} \implies t = \frac{\mathbf{a} \cdot \mathbf{x}}{\mathbf{a} \cdot \mathbf{a}} = (\mathbf{a}^T\mathbf{a})^{-1}\mathbf{a}^T\mathbf{x}
$$

**Task 1: Find the projection of $\mathbf{y}$ onto $\mathbf{x}$**

Data:
* $\mathbf{x} = (2, -1, 3)$
* $\mathbf{y} = (4, -1, 2)$.

Hint:
* Use `np.array(...)` to store a vector
* `np.inner(a, b)` computes the inner product of vector `a` and vector `b`
* The projection of $\mathbf{x}$ onto $\mathbf{a}$ is given by:
$
\quad \mathbf{p} = \dfrac{\mathbf{a} \cdot \mathbf{x}}{\mathbf{a} \cdot \mathbf{a}}\mathbf{a}
$

```python
# Import NumPy library for numerical computations and vector operations
import numpy as np

# Step 1.1: Define the original vectors x and y (given in the example)
x = np.array([2, -1, 3])   # Vector x = (2, -1, 3)
y = np.array([4, -1, 2])   # Vector y = (4, -1, 2)

# Step 1.2: Calculate the inner (dot) product of y and x: y · x
yx = np.inner(y, x)

# Step 1.3: Calculate the inner (dot) product of x with itself: x · x
xx = np.inner(x, x)

# Step 1.4: Compute the projection vector p of y onto x
p = (yx / xx) * x

# Print the final projection result
print("Projection vector p of y onto x =", p)
```

# Question 2: Finding a least-squares line for data

**Task 2: Find a least-squares solution of $A\mathbf{u} = \mathbf{y}$ from the given data.**

$$
A = \begin{bmatrix} 1 & 0 \\ 1 & 1 \\ 1 & 2 \\ 1 & 3 \end{bmatrix}, \quad
\mathbf{u} = \begin{bmatrix} a \\ b \end{bmatrix}, \quad
\mathbf{y} = \begin{bmatrix} 1 \\ 3 \\ 4 \\ 4 \end{bmatrix}
$$

Hint:
* The least-squares solution of $A\mathbf{u} = \mathbf{y}$ is given by $$
\mathbf{\hat u} = (A^T A)^{-1} A^T \mathbf{y}
$$
which minimizes $\|A\mathbf{u} - \mathbf{y}\|$.
* `X @ Y` can be used as the matrix mutliplication of matrices `X` and `Y`
* `np.linalg.inv(X)` can be used to find matrix inverse of matrix `X`


```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Step 2.1: Define given matrix A and vector y
# --------------------------
# Design matrix A from the linear regression problem
A = np.array([[1, 0], [1, 1], [1, 2], [1, 3]])

# Target output vector y
y = np.array([1, 3, 4, 4])

# --------------------------
# Step 2.2: Compute the least-squares solution
# --------------------------
# Step 2.2.1: Compute A transpose multiplied by A
ATA = A.T @ A

# Step 2.2.2: Compute inverse of ATA
ATA_inv = np.linalg.inv(ATA)

# Step 2.2.3: Compute A transpose multiplied by y
ATy = A.T @ y

# Step 2.2.4: Calculate the least-squares solution vector û (contains a and b)
u_hat = ATA_inv @ ATy

# Step 2.2.5: Extract intercept (a) and slope (b) from the solution
intercept = u_hat[0]
slope = u_hat[1]

# --------------------------
# Step 2.3: Print the result
# --------------------------
print("Least-squares solution u* =", u_hat)
print(f"Fitted line equation: y = {intercept:.1f} + {slope:.1f}x")

# --------------------------
# Step 2.4: Plot the data and the fitted line (matches TikZ figure)
# --------------------------
# Step 2.4.1: Extract x values from the first column of data
x_data = np.array([0, 1, 2, 3])
y_data = y

# Step 2.4.2: Create x values for plotting the fitted line
x_line = np.linspace(-0.5, 4, 100)
y_line = intercept + slope * x_line

# Step 2.4.3: Plot the figure below
plt.figure(figsize=(8, 6))                                            # Create figure with matching size
plt.scatter(x_data, y_data, color='red', s=60, zorder=5)              # Plot red data points (exact coordinates)
plt.plot(x_line, y_line, color='blue', linewidth=2)                   # Plot blue thick fitted line
plt.axhline(0, color='black', linewidth=1)                            # Draw centered axes
# plt.axvline(0, color='black', linewidth=1)
plt.xlabel(r'$x$', fontsize=12)                                       # Set labels and axis limits (exactly like LaTeX)
plt.ylabel(r'$y$', fontsize=12)
plt.xlim(-0.5, 4)
plt.ylim(0, 5)
plt.title("Fig. 4: Least-squares line fitting the data", fontsize=12) # Add title (figure caption)
plt.xticks()                                                          # Show grid and ticks
plt.yticks()
plt.tight_layout()
plt.show()
```

# Question 3: Finding a suitable curve for data (Curve Fitting)


**Task 3: Find the best fit curve $u = a + bx + cx^2$**

Data:

$$
(0,1),\ (1,3),\ (2,4),\ (3,4)
$$

Hint:
* The fact that all 4 points of data should satisfy the quadratic equation $u = a + bx + cx^2$ can be used to form the design matrix $A$ and vector $y$
* The least-squares solution of $A\mathbf{u} = \mathbf{y}$ is given by $$
\mathbf{\hat u} = (A^T A)^{-1} A^T \mathbf{y}
$$
which minimizes $\|A\mathbf{u} - \mathbf{y}\|$.

```python
import numpy as np
import matplotlib.pyplot as plt

# =============================================================================
# Step 3.1: Define the given data points (x, u)
# =============================================================================
x_data = np.array([0, 1, 2, 3])
u_data = np.array([1, 3, 4, 4])

# =============================================================================
# Step 3.2: Build the design matrix A for QUADRATIC fitting (u = a + bx + cx²)
# =============================================================================
A = np.array([
    [1, 0, 0],   # x=0 → 1, 0, 0²
    [1, 1, 1],   # x=1 → 1, 1, 1²
    [1, 2, 4],   # x=2 → 1, 2, 2²
    [1, 3, 9]    # x=3 → 1, 3, 3²
])

# Target vector y (observed u values)
y = np.array([1, 3, 4, 4])

# =============================================================================
# Step 3.3: Compute the least-squares solution
# =============================================================================
# Compute A transpose multiplied by A
ATA = A.T @ A

# Compute inverse of ATA
ATA_inv = np.linalg.inv(ATA)

# Compute A transpose multiplied by y
ATy = A.T @ y

# Solve for coefficients [a, b, c]
u_hat = ATA_inv @ ATy

# Extract the quadratic coefficients
a, b, c = u_hat[0], u_hat[1], u_hat[2]

# =============================================================================
# Step 3.4: Print the results
# =============================================================================
print("Least-squares solution u* =", u_hat)
print(f"Quadratic curve: u = {a:.1f} + {b:.1f}x + {c:.1f}x²")

# =============================================================================
# Step 3.5: Plot the data points and the fitted quadratic curve
# (Matches the LaTeX TikZ figure exactly)
# =============================================================================
x_curve = np.linspace(-0.5, 4, 100)                                               # Create smooth x values for plotting the curve
u_curve = a + b * x_curve + c * x_curve**2                                        # Compute u values using the quadratic formula
plt.figure(figsize=(8, 6))                                                        # Create figure with matching size
plt.scatter(x_data, u_data, color='red', s=60, zorder=5)                          # Plot red data points (solid circles)
plt.plot(x_curve, u_curve, color='blue', linewidth=2)                             # Plot blue thick quadratic fitting curve
plt.axhline(0, color='black', linewidth=1)                                        # Draw centered x and y axes
# plt.axvline(0, color='black', linewidth=1)
plt.xlabel(r'$x$', fontsize=12)                                                   # Set axis labels (math notation)
plt.ylabel(r'$u$', fontsize=12)
plt.xlim(-0.5, 4)                                                                 # Set axis limits (exactly like LaTeX)
plt.ylim(0, 5)
plt.title("Fig. 5: Quadratic least-squares curve fitting the data", fontsize=12)  # Add figure title (matches LaTeX caption)
plt.tight_layout()                                                                # Clean layout and show plot
plt.show()
```
