# Expectation, Variance, Standard Deviation

## Example 1: Discrete Random Variable $X$

**Task 1.1: Find the Expecation, Variance and Standard Deviation of a discrete random variable**

To do: replace `...` below to complete the tasks

Given PMF table for discrete random variable $X$:

$$
\begin{array}{c|cccc|c}
x & 0 & 1 & 2 & 3 & \textrm{Sum} \\
\hline
 P(X=x)  &  0.010  &  0.840  &  0.145  &  0.005  &  1
\end{array}
$$

Find the Expecation, Variance and Standard Deviation of $X$.

**Hints:**
- Expectation:
    $$
    \mathbb{E}[X]  = \sum_{i=1}^n x_i p_i
    $$
    
-  Variance:
    $$
    \operatorname{Var}(X) = \mathbb{E}[X^2] - \big(\mathbb{E}[X]\big)^2,\quad \mathbb{E}[X^2]=\sum_{i=1}^n x_i^2 p_i
    $$
    
-  Standard deviation:
    $$
    \operatorname{SD}(X) = \sqrt{\operatorname{Var}(X)}
    $$

- Use lists and for loops to compute summation
- `zip(list1, list2)` combine elements from both lists `list1` and `list2` into a sequence of tuples.
- `math.sqrt(a)` returns the value of $\sqrt a$
- `a ** n` returns the value of $a^n$

```python
import math

# Task 1.1.1: Define random variable values and probabilities
x_vals = [0, 1, 2, 3]
p_vals = [0.010, 0.840, 0.145, 0.005]

# Task 1.1.2: Compute the Expectation E[X]
E_X = sum(x * p for x, p in zip(x_vals, p_vals))

# Task 1.1.3: Compute E[X^2]
E_X2 = sum((x**2) * p for x, p in zip(x_vals, p_vals))

# Task 1.1.4: Compute the Variance Var(X)
Var_X = E_X2 - E_X**2

# Task 1.1.5: Compute the Standard deviation SD(X)
SD_X = math.sqrt(Var_X)

# Task 1.1.6: Print the results
print(f"E(X) = {E_X:.7f}")
print(f"Var(X) = {Var_X:.7f}")
print(f"SD(X) = {SD_X:.7f}")
```

**Task 1.2: Plot the graph of the PMF of the random varaible**

The plot visualizes the probability distribution of $X$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for X from Task 1.1.1
x_vals = [0, 1, 2, 3]
p_vals = [0.010, 0.840, 0.145, 0.005]

# Create the bar plot for the PMF
plt.figure(figsize=(8, 5))
plt.bar(x_vals, p_vals, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('X (Number of blue balls)')
plt.ylabel('P(X=x)')
plt.title('Probability Mass Function (PMF) of X')

# Set x-axis ticks to be the x_vals
plt.xticks(x_vals)

# Set y-axis limits to start from 0 and go slightly above max probability
plt.ylim(0, max(p_vals) * 1.1)

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add text labels on top of each bar
for i, prob in enumerate(p_vals):
    plt.text(x_vals[i], prob + 0.01, f'{prob:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

##  Example 2: Continuous Random Variable $X$

**Task 2.1: Find the Expecation and Variance of continuous  random variables**

To do: replace `...` below to complete the tasks

Let $X$ be a random variable with PDF $f(t)=3t^2,\ 0\le t\le1$. \
Define $Y=4X+2$. Compute $\mathbb{E}[X],\operatorname{Var}(X),\mathbb{E}[Y],\operatorname{Var}(Y)$.

**Hints:**
+ Expectation:
    $$
    \mathbb{E}[X] = \int_{-\infty}^{\infty} t f_X(t) dt
    $$
    
+  Variance:
    $$
    \operatorname{Var}(X) = \mathbb{E}\big[(X-\mu_X)^2\big] = \int_{-\infty}^{\infty} (t-\mu_X)^2 f_X(t)dt
    = \mathbb{E}[X^2] - (\mathbb{E}[X])^2,\quad \mathbb{E}[X^2]=\int_{-\infty}^\infty t^2 f_X(t)dt
    $$
- `sp.Symbol('x', real=True)` declares `x` as a symbolic variable with real values \
(documentation: https://docs.sympy.org/latest/index.html)
- `sp.integrate(f, (x, a, b))` returns the definite integral $\int_a^b f(x) dx$ \
(documentation: https://docs.sympy.org/latest/modules/integrals/integrals.html)


```python
import sympy as sp

# Task 2.1.1: Define symbolic variable and PDF function
t = sp.Symbol('t', real=True)
f = 3 * t**2

# Task 2.1.2: Compute E[X], E[X^2] and Var(X)
E_X = sp.integrate(t * f, (t, 0, 1))
E_X2 = sp.integrate(t**2 * f, (t, 0, 1))
Var_X = E_X2 - E_X**2

# Task 2.1.3: Compute E[Y], E[Y^2]
Y_expr = 4*t + 2      # Y = 4X + 2
E_Y = sp.integrate(Y_expr * f, (t, 0, 1))
E_Y2 = sp.integrate(Y_expr**2 * f, (t, 0, 1))
Var_Y = E_Y2 - E_Y**2

# Print exact symbolic and decimal values
print("=== X with PDF: f(t) = 3t², 0 ≤ t ≤ 1 ===")
print(f"E[X] = {E_X}")
print(f"Var(X) = {Var_X}")
print("\n=== Y = 4X + 2 ===")
print(f"E[Y] = {E_Y}")
print(f"Var(Y) = {Var_Y}")
```

**Task 2.2: Plot the graph of the PDF of the random varaibles**

The plot visualizes the probability distributions of $X$ and $Y$.

## Example 3: Normal Distribution and Uniform Distribution

**Task 3.1.1: Properties of Normal Distribution**

To do: replace `...` below to complete the tasks

Given a Normal random variable $X \sim N(\mu, \sigma^2)$ with parameters $\mu$ and $\sigma^2$ with PDF

$$
f_X (x) = \frac{1}{\sqrt {2 \pi \sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

Compute the following quantities:
1. (valid PDF check) $\int_{-\infty}^\infty f(x)dx$

2. $\mathbb{E}[X]=\int_{-\infty}^\infty  x f(x)dx$

3. $\mathbb{E}[X^2]=\int_{-\infty}^\infty  x^2 f(x)dx$

4. $\operatorname{Var}(X)=\mathbb{E}[X^2]-(\mathbb{E}[X])^2$

5. $\sigma_X=\sqrt{\operatorname{Var}(X)}$

in terms of the parameters.

**Hints:**
- `sp.Symbol('x', real=True)` declares `x` as a symbolic variable with real values

- `sp.integrate(f, (x, a, b))` returns the definite integral $\int_a^b f(x) dx$ \
(documentation: https://docs.sympy.org/latest/modules/integrals/integrals.html)

- `sp.sqrt(x)` represents $\sqrt x$

- `sp.pi` representes $\pi$
- `sp.oo` represents $\infty$ \
(documentation: https://docs.sympy.org/latest/modules/core.html)



```python
import sympy as sp

# Task 3.1.1.0: Define symbolic variables and PDF parameters
x = sp.Symbol("x", real=True)
mu = sp.Symbol("mu", real=True)
sigma = sp.Symbol("sigma", positive=True)

# Define Normal PDF f(x)
f_x = (1 / (sigma * sp.sqrt(2 * sp.pi))) * sp.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

print("=== Defined Normal PDF f(x) ===")
display(f_x)
```

```python
# Task 3.1.1.1: Verify PDF validity
integral_f = sp.integrate(f_x, (x, -sp.oo, sp.oo))

print("1: Integral of f(x) from -∞ to +∞")
display(integral_f)
```

```python
# Task 3.1.1.2: Compute the expected value E[X] = ∫x*f(x) dx
E_X = sp.integrate(x * f_x, (x, -sp.oo, sp.oo))

print("2: E[X] = ∫x f(x) dx")
```

```python
# Task 3.1.1.3: Compute the second moment E[X²] = ∫x²*f(x) dx
E_X2 = sp.integrate((x ** 2) * f_x, (x, -sp.oo, sp.oo))

print("3: E[X²] = ∫x² f(x) dx")
display(E_X2)
```

```python
# Task 3.1.1.4: Compute the variance Var(X) = E[X²] - (E[X])²
Var_X = E_X2 - (E_X) ** 2

print("4: Var(X) = E[X²] - (E[X])²")
display(sp.simplify(Var_X))
```

```python
# Task 3.1.1.5: Compute the standard deviation σ_X = sqrt(Var(X))
sigma_X = sp.sqrt(Var_X)

print("5: σ_X = sqrt(Var(X))")
display(sp.simplify(sigma_X))
```

**Task 3.1.2: Plot the graph of Normal Distribution**

The plot visualizes a normal probability distribution. Adjust the sliders for parameters to see the changes.

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Define the PDF for a normal distribution
def normal_pdf(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Create a function to plot the normal distribution
def plot_normal_distribution(mu, sigma_squared):
    if sigma_squared <= 0:
        print("Error: Variance (sigma_squared) must be greater than 0.")
        return

    sigma = np.sqrt(sigma_squared)

    fig, ax = plt.subplots(figsize=(8, 4))

    # Fixed x-limits for the plot
    x_plot_min = -10
    x_plot_max = 10
    x_vals = np.linspace(x_plot_min, x_plot_max, 500)

    y_vals = normal_pdf(x_vals, mu, sigma)

    ax.plot(x_vals, y_vals, label=f'N(μ={mu:.2f}, σ²={sigma_squared:.2f})')
    ax.fill_between(x_vals, y_vals, alpha=0.2)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title('Normal Probability Distribution')

    # Fixed y-limits for the plot
    ax.set_ylim(-0.05, 1.0)
    ax.set_xlim(-10, 10)

    ax.legend()
    ax.grid(True)

    # Add label for the height of the PDF
    pdf_height = normal_pdf(mu, mu, sigma) # Height at the mean
    ax.text(mu, pdf_height * 1.05, f'Height = {pdf_height:.2f}',
            horizontalalignment='center', verticalalignment='bottom', fontsize=10, color='red')

    plt.tight_layout() # Adjust layout to prevent labels from being cut off
    plt.show()

# Create sliders for mu and sigma_squared
mu_slider = FloatSlider(min=-5, max=5, step=0.1, value=0, description='mu')
sigma_squared_slider = FloatSlider(min=0.2, max=5, step=0.1, value=1, description='sigma^2')

# Use interact to link the plot function to the sliders
print("Adjust the sliders to see the Normal Distribution change:")
interact(plot_normal_distribution, mu=mu_slider, sigma_squared=sigma_squared_slider);
```

**Task 3.2.2: Plot the graph of Uniform Distribution**

The plot visualizes a uniform probability distribution. Adjust the sliders for parameters to see the changes.

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Define the PDF for a uniform distribution
def uniform_pdf(x, a, b):
    if a >= b:
        return np.zeros_like(x) # Handle invalid range
    return np.where((x >= a) & (x <= b), 1 / (b - a), 0)

# Create a function to plot the uniform distribution
def plot_uniform_distribution(a, b):
    if a >= b:
        print("Error: 'a' must be less than 'b'. Please adjust the sliders.")
        return

    fig, ax = plt.subplots(figsize=(8, 4))
    x_vals = np.linspace(-2.95, 2.95, 500)

    y_vals = uniform_pdf(x_vals, a, b)

    ax.plot(x_vals, y_vals, label=f'Uniform PDF f(x) on [{a:.2f}, {b:.2f}]')
    ax.fill_between(x_vals, y_vals, alpha=0.2)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title('Uniform Probability Distribution')

    # Calculate PDF height
    pdf_height = 1 / (b - a)
    ax.set_ylim(0, pdf_height * 1.2) # Set y-limit based on PDF height with buffer

    ax.legend()
    ax.grid(True)

    # Add label for the height of the PDF
    # Position: x at midpoint of [a,b], y slightly above the pdf_height
    ax.text((a + b) / 2, pdf_height * 1.05, f'Height = {pdf_height:.2f}',
            horizontalalignment='center', verticalalignment='bottom', fontsize=10, color='red')

    plt.xlim(-3,3)
    plt.ylim(-0.05, 3)
    plt.tight_layout() # Adjust layout to prevent labels from being cut off
    plt.show()

# Create sliders for a and b
a_slider = FloatSlider(min=-3, max=3, step=0.1, value=0, description='a')
b_slider = FloatSlider(min=-3, max=3, step=0.1, value=1, description='b')

# Use interact to link the plot function to the sliders
print("Adjust the sliders to see the uniform distribution change:")
interact(plot_uniform_distribution, a=a_slider, b=b_slider);
```

# Joint Probability Mass Function for Discrete Random Variables

## Example 4: Hypergeometric Draw of Colored Balls

**Task 4.1: Properties of Joint Probability Distribution**

To do: replace `...` below to complete the tasks

A pocket contains

- $3$ blue balls, $2$ red balls, $3$ green balls; total $3+2+3=8$ balls.

Draw $2$ balls **without replacement**.

Define:

- $X$ = number of blue balls drawn

- $Y$ = number of red balls drawn

Tasks:

1. Derive formula for joint PMF $p(x,y)={P}(X=x,Y=y)$.

2. Construct the full joint probability table.

3. Compute ${P}(X+Y\le 1)$.

4. Find the marginal distribution of $X$.

5. Find the marginal distribution of $Y$.

**Hints:**
- `math.comb(n,k)` returns the value of ${}_nC_k$
- Define an extended combination function `comb(n,k)` which returns ${}_nC_k$ if $0\le k \le n$ and reutrn $0$ otherwise.
- Use a dictionary `{key:value}`to store the probability values $p(x,y)$ with key `(x,y)`.
- The $\textbf{joint probability mass function}$ is defined:
$$
p_{ij}=p(x_i,y_j) = \mathbb{P}(X=x_i,\ Y=y_j),\quad x_i,y_j\in\mathbb{R},\ i=1,\dots,s,\ j=1,\dots,t
$$

- The $\textbf{marginal distribution}$ is defined:

$$
\mathbb{P}(X=x_i) = \sum_{j=1}^t \mathbb{P}(X=x_i,Y=y_j) = \sum_{j=1}^t p_{ij}
$$
$$
\mathbb{P}(Y=y_j) = \sum_{i=1}^s \mathbb{P}(X=x_i,Y=y_j) = \sum_{i=1}^s p_{ij}
$$

```python
import math
from fractions import Fraction

# Define the extended combination function
def comb(n, k):
    if k < 0 or k > n:
        return 0
    return math.comb(n, k)

# Parameters
n_blue = 3
n_red = 2
n_green = 3
total = n_blue + n_red + n_green
draw = 2

# Task 4.1.1: Compute all joint probabilities p(x,y)
denominator = comb(total, draw)     # Total possible ways to draw 2 balls without replacement
joint_probs = {}                    # Initialize dictionary to store probability values
for x in [0,1,2]:
    for y in [0,1,2]:
        z = draw - x - y
        numerator = comb(n_blue, x) * comb(n_red, y) * comb(n_green, z)     # Total desired ways to draw 2 balls
        prob = Fraction(numerator, denominator)
        joint_probs[(x,y)] = prob

# Print joint table
print("=== Joint Probability Table p(x,y) ===")
for x in [0,1,2]:
    row = []
    for y in [0,1,2]:
        row.append(str(joint_probs[(x,y)]))
    print(f"x = {x}: y0 = {row[0]}, y1 = {row[1]}, y2 = {row[2]}")
```

```python
# Task 4.1.2: Compute P(X+Y ≤ 1)
p_sum_le1 = joint_probs[(0,0)] + joint_probs[(0,1)] + joint_probs[(1,0)]
print(f"P(X+Y ≤ 1) = {p_sum_le1}")
```

```python
# Task 4.1.3: Marginal Distirbution of X
marginal_x_probs = {}       # Initialize dictionary to store probability values
for x in [0,1,2]:
    s = sum(joint_probs[(x,y)] for y in [0,1,2])
    marginal_x_probs[x] = s
print("=== Marginal Distribution of X ===")
for x, val in marginal_x_probs.items():
    print(f"P(X = {x}) = {val}")
```

```python
# Task 4.1.4: Marginal Distribution of Y
marginal_y_probs = {}     # Initialize dictionary to store probability values
for y in [0,1,2]:
    s = sum(joint_probs[(x,y)] for x in [0,1,2])
    marginal_y_probs[y] = s
print("=== Marginal Distribution Y ===")
for y, val in marginal_y_probs.items():
    print(f"P(Y={y}) = {val}")
```

**Task 4.2: Plot the Joint PMF and Marginal PMFs**

# Joint Probability Density Function for Continuous Random Variables

## Example 5: Joint PDF

**Task 5.1: Find the marginal PDF of random varaibles**

To do: replace `...` below to complete the tasks

The joint PDF of continuous random variables $X,Y$ is
$$
f(x,y) =
\begin{cases}
x+y & 0<x<1,\ 0<y<1 \\
0 & \text{otherwise}
\end{cases}
$$
1. Calculate the marginal density functions $f_X(x)$ and $f_Y(y)$ over the support region $(0,1)\times(0,1)$.

2. Verify that $\displaystyle\int_{-\infty}^\infty f_X(x) dx = \int_{-\infty}^\infty f_Y(y) dy = 1$

**Hints:**
- Definition of marginal PDF:
$$
f_X(x) = \int_{-\infty}^{\infty} f(x,y)\,dy
$$
- `sp.Symbol('x')` declares `x` as a symbolic variable
- `sp.Piecewise( (expr,cond), (expr,cond))` represents a piecewise function with expression `expr` and condition `cond`
- `sp.integrate(f, (x, a, b))` returns the definite integral $\int_a^b f(x) dx$ \
(documentation: https://docs.sympy.org/latest/modules/integrals/integrals.html)
- `sp.oo` represents $\infty$ \
(documentation: https://docs.sympy.org/latest/modules/core.html)

```python
import sympy as sp

# Task 5.1.1: Define symbolic variables
x, y = sp.Symbol('x'), sp.Symbol('y')

# Task 5.1.2: Define the joint PDF
f = sp.Piecewise((x + y, (0 < x) & (x < 1) & (0 < y) & (y < 1)), (0, True))

# Task 5.1.3: Compute marginal PDF f_X(x) = integrate f w.r.t y from 0 to 1
f_X = sp.integrate(f, (y, -sp.oo, sp.oo))

# Task 5.1.4: Compute marginal PDF f_Y(y) = integrate f w.r.t x from 0 to 1
f_Y = sp.integrate(f, (x, -sp.oo, sp.oo))

# Print symbolic results
print("Marginal PDF f_X(x) =")
display(f_X)
print("\nMarginal PDF f_Y(y) =")
display(f_Y)

# Task 5.1.5: Verify normalization integral for f_X and f_Y
norm_X = sp.integrate(f_X, (x, -sp.oo, sp.oo))
norm_Y = sp.integrate(f_Y, (y, -sp.oo, sp.oo))

print("\nIntegral of f_X(x) over [0,1] =", norm_X)
print("Integral of f_Y(y) over [0,1] =", norm_Y)
```

**Task 5.2: Plotting the Joint PDF and Marginal PDFs**

This plot visualizes the Joint PDF $f(x,y)$ and the marginal PDFs $f_X(x)$ and $f_Y(y)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, IntSlider
from IPython.display import display

# Define the joint PDF function
def joint_pdf(x, y):
    return np.where((x > 0) & (x < 1) & (y > 0) & (y < 1), x + y, 0)

# Create a meshgrid for the 3D plot
x_vals = np.linspace(0, 1, 100)
y_vals = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = joint_pdf(X, Y)

def plot_joint_pdf_interactive(elev=30, azim=45):
    # Create the 3D plot
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x,y)')
    ax.set_title('Joint Probability Density Function f(x,y) = x+y')

    # Set the viewing angle
    ax.view_init(elev=elev, azim=azim)

    plt.show()

print("Adjust the sliders to change the viewing angle of the Joint PDF:")
interact(plot_joint_pdf_interactive,
         elev=IntSlider(min=0, max=90, step=5, value=30, description='Elevation'),
         azim=IntSlider(min=0, max=360, step=5, value=225, description='Azimuth'));
```

# Covariance Matrix

## Example 6: Sample Covariance Matrix Calculation

**Task 6: Find the Sample Covariance Matrix**

To do: replace `...` below to complete the tasks

Given 5 variables ($X_1,X_2,X_3,X_4,X_5$), $n=6$ sample observations per variable:


$$
\begin{array}{c|cccccc}
Variable & Obs1 & Obs2 & Obs3 & Obs4 & Obs5 & Obs6 \\
\hline
X_1 & 1 & 2 & 3 & 4 & 5 & 6 \\
X_2 & 2 & 3 & 5 & 6 & 1 & 9 \\
X_3 & 3 & 5 & 5 & 5 & 10 & 8 \\
X_4 & 10 & 20 & 30 & 40 & 50 & 55 \\
X_5 & 7 & 8 & 9 & 4 & 6 & 10
\end{array}
$$

Find the Sample Covariance Matrix $\widehat{\boldsymbol{\Sigma}} = [\widehat{\operatorname{Cov}}(X_i,X_j)]_{1 \le i, j \le 5}$.

**Hints:**
- Sample covariance formula (unbiased, denominator $n-1$):
$$
\widehat{\operatorname{Cov}}(X_i,X_j)=\frac{1}{n-1}\sum_{k=1}^n \big(X_{i,k}-\bar{X}_i\big)\big(X_{j,k}-\bar{X}_j\big),\quad
\widehat{\operatorname{Var}}(X_i)=\widehat{\operatorname{Cov}}(X_i,X_i)
$$
- Use `np.array()` to store the data
- `np.cov(data, rowvar=True, ddof=1)` returns the **Sample** Covaraince Matrix of the data array `data` where rows of `data` represent variables and colums represent observations

```python
import numpy as np

# Step 1: Define dataset matrix (rows = variables X_1, X_2, X_3, X_4, X_5; columns = observations)
data = np.array([
    [1, 2, 3, 4, 5, 6],       # X_1
    [2, 3, 5, 6, 1, 9],       # X_2
    [3, 5, 5, 5, 10, 8],      # X_3
    [10, 20, 30, 40, 50, 55], # X_4
    [7, 8, 9, 4, 6, 10]       # X_5
])

# Step 2: NumPy built-in cov function (rowvar=True: rows = variables, ddof=1 for unbiased n-1)
cov_matrix = np.cov(data, rowvar=True, ddof=1)

# Print formatted 7-decimal output matching Sage
print("Sample Covariance Matrix (unbiased, ddof=1):")
print(np.round(cov_matrix, 7))
```
